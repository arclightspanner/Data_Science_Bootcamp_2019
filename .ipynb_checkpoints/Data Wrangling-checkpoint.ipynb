```python
import glob
import pandas as pd
import os
```

```python
os.chdir("C:/Users/alanl/Desktop/Repository/Data_Science_Bootcamp_2019/collected data9.6.2019")
extension = 'csv'
all_filenames = [i for i in glob.glob('player_*.{}'.format(extension))]
```

```python
print(all_filenames)
```

```python
#combine all files in the list
df = pd.concat([pd.read_csv(f) for f in all_filenames ])
```

```python
# seems like the indexing is incorrect
df.info()
```

```python
# We need to reset the index
df = df.reset_index(drop=True)
df.tail()
```

```python
df.describe()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df_info = df.describe()
```

```python
# if the max value in column is the same as the min value then the column should be dropped
for i in df_info.columns:
    if df_info[i]['min'] == df_info[i]['max']:
        df = df.drop([i], axis=1)
    else:
        continue
```

```python
# check if the naming of player data contains duplicates, in this scenario, we can see that xSentineIx showed up twice
df['name'].value_counts(dropna=False).head()
```

```python
# we can see that this player shows up twice and it has the same values
df[df['name']=='xSentineIx']
```

```python
# drop duplicate row with index 36978 and reset index again
df = df.drop(36978)
df = df.reset_index(drop=True)
```

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Describe the column
df['bestStreak'].describe()
```

```python
streak_data = pd.DataFrame(df['bestStreak'].value_counts().sort_index())
```

```python
# Plot the frequency chart, we can see that streak data is very extreme, with the 25% and 75% value being the same.
# we can remove columns where the 25% is the same as 75%
_ = plt.plot(streak_data.index,streak_data.bestStreak,color="dodgerblue")
_ = plt.xlabel('bestStreak Numbers')
_ = plt.ylabel('number of players')
_ = plt.title('bestStreak player count')
_ = plt.show()
```

```python
df_info = df.describe()
for i in df_info.columns:
    if df_info[i]['25%'] == df_info[i]['75%']:
        df = df.drop([i], axis=1)
    else:
        continue
```

```python
# import player ranking obtained through webscraping and attempt to merge ranking data with stat data df
df2 = pd.read_csv("C:/Users/alanl/Desktop/Repository/Data_Science_Bootcamp_2019/players_full_dataset.csv")
df2.head()
```

```python
merged_data = df2.merge(df, left_on=' player_id', right_on='name')
merged_data.tail()
```

```python
# notice that player_id column name has white space before it, so we should remove all white space at the beginning of column 
# names
merged_data.columns = merged_data.columns.str.replace('^ +', '')
```

```python
# test if value 999.99 exists in the dataframe, if nothing is printed then there are no missing values;
for i in merged_data.columns:
    true_false = set(merged_data[i]==999.99)
    if True in true_false:
        print(i)
```

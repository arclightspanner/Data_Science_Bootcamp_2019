## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
```

```python
import json
from pandas.io.json import json_normalize
```

```python
# load as Pandas dataframe
json_df = pd.read_json('data/world_bank_projects.json')
json_df
```

```python
# Find the 10 countries with most projects
df_count = json_df['countryname'].value_counts()
df_count_top_10 = df_count.head(10)
df_count_top_10
print('10 Countries with Most Projects:')
print(df_count_top_10.head(10))
```

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')
# Look at the content in mjtheme_namecode
df_count = json_df['mjtheme_namecode']
df_count
```

```python
# Convert the object to one json file
converted_json = json.loads(json_df.mjtheme_namecode.to_json(orient='values'))

# Break down the json list using chain from the itertools library
from itertools import chain
break_down_list = list(chain.from_iterable(converted_json))
break_down_list[:10]
```

```python
break_down_df = pd.DataFrame(break_down_list)
project_count = break_down_df['name'].value_counts()
print('10 Major Project themes:')
print(project_count.head(10))
```

```python
# Above you will notice that some entries have only the code and the name is missing. 
# Create a dataframe with the missing names filled in.

# Iterate over DataFrame rows as namedtuples.
for row in break_down_df.itertuples():
    print(row)
```

```python
# Create a blank dictionary and map all the non-blank entries into the dictionary
code_dict = {}
for row in break_down_df.itertuples():
        if row[2] != '':
            code_dict[row[1]] = row[2]
code_dict
```

```python
# Fill in missing theme names with code dictionary
for row in break_down_df.itertuples():
    if row[2] == '':
        break_down_df.set_value(row[0], 'name', code_dict[row[1]])
```

```python
break_down_df
```

```python
# Check if there are still any missing values
var = break_down_df.isnull().values.any()
var = True
print('There are no missing values')
if not var:
        print('There are some missing values')
```

```python
italy.plot.scatter('year','gdp_per_day', logy=True, ylim=(10**0, 10**2.5));
```

```python
italy.plot.scatter('gdp_per_day','life_expectancy',logx = True, xlim=(10**0, 10**2.5));
```

```python
# if the year is an integer multiple of 10, use a large dot, otherwise use a small dot
size = np.where(italy.year % 10 ==0,110,2)

italy.plot.scatter('gdp_per_day','life_expectancy',logx = True, s = size, xlim=(10**0.4, 10**2.5));
```

```python
data = gapminder.query('(country == "Italy") or (country == "United States")')
```

```python
size = np.where(data.year % 10 == 0, 130, 2)
color = np.where(data.country == 'Italy', 'blue','orange')
```

```python
data.plot.scatter('gdp_per_day','life_expectancy', logx = True, s=size,c=color, xlim=(10**0.4,10**2.5))
```

```python
ax = data.plot.scatter('gdp_per_day','life_expectancy', logx = True, s=size,c=color)
data[data.country=='Italy'].plot.line(x='gdp_per_day',y='life_expectancy',ax=ax,xlim=(10**0.4,10**2.5))
```
